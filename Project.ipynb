# Deal with the missing values

```python
df.isnull().sum()
```

```python
df["LoanAmount"]=df["LoanAmount"].replace(np.NaN,df["LoanAmount"].mean())
```

```python
print(df["LoanAmount"])
```

```python
df.isnull().sum()
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.NaN,df["Loan_Amount_Term"].mean())
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.NaN,df["Credit_History"].mean())
```

```python
df["Gender"].mode()
```

```python
df["Gender"]=df["Gender"].replace(np.NaN,"Male")
```

```python
df["Married"].mode()
```

```python
df["Married"]=df["Married"].replace(np.NaN,"Yes")
```

```python
df["Dependents"].mode()
```

```python
df["Dependents"]=df["Dependents"].replace(np.NaN,"0")
```

```python
df["Self_Employed"].mode()
```

```python
df["Self_Employed"]=df["Self_Employed"].replace(np.NaN,"No")
```

# Machine learning model




```python
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
```

```python
model=SVC()
enc=LabelEncoder()
```

```python
df.Loan_ID=enc.fit_transform(df.Loan_ID)
df.Gender=enc.fit_transform(df.Gender)
df.Married=enc.fit_transform(df.Married)
df.Dependents=enc.fit_transform(df.Dependents)
df.Education=enc.fit_transform(df.Education)
df.Self_Employed=enc.fit_transform(df.Self_Employed)
df.Property_Area=enc.fit_transform(df.Property_Area)
```

```python
df.drop("Loan_ID",axis=1)
```

for i in df.Loan_ID:
